```python
import pandas as pd
```

```python
# -------------------------------
# IMPORT DATA
# -------------------------------
data = pd.read_csv("data.csv", sep=';')

print("Imported Data:")
print(data.head())

# -------------------------------
# DATA PROCESSING
# -------------------------------
data['Total'] = (
    data['Curricular units 1st sem (grade)'] +
    data['Curricular units 2nd sem (grade)']
)

data['Average'] = data['Total'] / 2

print("\nProcessed Data:")
print(data[['Total', 'Average']].head())

# -------------------------------
# EXPORT DATA
# -------------------------------
data.to_csv("processed_student_data.csv", index=False)

print("\nData successfully exported to 'processed_student_data.csv'")
```
