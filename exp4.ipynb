```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data.csv", sep=';')

print("Dataset Preview:")
print(data.head())
```

```python
plt.figure(figsize=(6,4))
plt.hist(data['Unemployment rate'], bins=10)
plt.title("Histogram of Unemployment Rate")
plt.xlabel("Unemployment Rate")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6,4))
data['Target'].value_counts().plot(kind='bar')
plt.title("Count of Target Categories")
plt.xlabel("Target")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))
plt.scatter(data['GDP'], data['Inflation rate'])
plt.title("GDP vs Inflation Rate")
plt.xlabel("GDP")
plt.ylabel("Inflation Rate")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=data['Curricular units 1st sem (grade)'])
plt.title("Boxplot of 1st Semester Grades")
plt.show()
```
